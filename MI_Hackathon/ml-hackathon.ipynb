```python
train = pd.read_csv("/kaggle/input/pesurrcampusmicompetitionb/train.csv")
```

```python
train_df = pd.DataFrame(train)
train_df.head()
```

```python
train_df = pd.DataFrame(train)
train_df.head()
```

```python
target_column = train_df['label']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_df.iloc[:, 1:-1] = sc.fit_transform(train_df.iloc[:, 1:-1])
```

```python
train_df.head()
```

```python
train_df = train_df.drop('length', axis = 1)
```

```python
train_df = train_df.drop(['mfcc10_var','tempo', 'mfcc17_mean','mfcc20_var', 'spectral_centroid_mean'] , axis = 1)
```

```python
cols = train_df.iloc[:, 1:-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cols, target_column, test_size = 0.2, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
test = pd.read_csv('/kaggle/input/pesurrcampusmicompetitionb/test.csv')
```

```python
test_df =pd.DataFrame(test)
test_df.head()
```

```python
id = test_df['id']
test_df = test_df.drop('id', axis = 1)
```

```python
test_df = test_df.drop(['mfcc10_var','tempo', 'mfcc17_mean','mfcc20_var', 'spectral_centroid_mean'] , axis = 1)
```

```python
test_df = test_df.drop('length', axis = 1)
test_df = sc.fit_transform(test_df)
```

```python
y_pred_test = classifier.predict(test_df)
```

```python
submission_df = pd.DataFrame()
submission_df['id'] = id
```

```python
submission_df['label'] = y_pred_test
```

```python
for i in range(0, len(submission_df)):
    if submission_df.iloc[i, 1] == 'blues':
       submission_df.iloc[i, 1] = 0
    elif submission_df.iloc[i, 1] == 'classical':
       submission_df.iloc[i, 1] = 1
    elif submission_df.iloc[i, 1] == 'country':
       submission_df.iloc[i, 1] = 2
    elif submission_df.iloc[i, 1] == 'disco':
       submission_df.iloc[i, 1] = 3
    elif submission_df.iloc[i, 1] == 'hiphop':
       submission_df.iloc[i, 1] = 4
    elif submission_df.iloc[i, 1] == 'jazz':
       submission_df.iloc[i, 1] = 5
    elif submission_df.iloc[i, 1] == 'metal':
       submission_df.iloc[i, 1] = 6
    elif submission_df.iloc[i, 1] == 'pop':
       submission_df.iloc[i, 1] = 7
    elif submission_df.iloc[i, 1] == 'reggae':
       submission_df.iloc[i, 1] = 8
    elif submission_df.iloc[i, 1] == 'rock':
       submission_df.iloc[i, 1] = 9
```

```python
submission_df.to_csv('music_genre_predictions.csv', index = False)
```
