```python
import matplotlib.pyplot as plt

train = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
test = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')
```

```python
train_df = pd.DataFrame(train)
train_df
```

```python
train_df = train_df.drop(['PassengerId', 'Name'], axis = 1)
```

```python
train_df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.NaN, strategy = 'mean')

train_df["Age"] = imputer.fit_transform(train_df[["Age"]])
train_df["RoomService"] = imputer.fit_transform(train_df[["RoomService"]])
train_df["FoodCourt"] = imputer.fit_transform(train_df[["FoodCourt"]])
train_df["ShoppingMall"] = imputer.fit_transform(train_df[["ShoppingMall"]])
train_df["Spa"] = imputer.fit_transform(train_df[["Spa"]])
train_df["VRDeck"] = imputer.fit_transform(train_df[["VRDeck"]])
```

```python
imputer2 = SimpleImputer(missing_values = np.NaN, strategy = 'most_frequent')
train_df["HomePlanet"] = imputer2.fit_transform(train_df[["HomePlanet"]])
train_df["Cabin"] = imputer2.fit_transform(train_df[["Cabin"]])
train_df["Destination"] = imputer2.fit_transform(train_df[["Destination"]])
train_df["VIP"] = imputer2.fit_transform(train_df[["VIP"]])
train_df["CryoSleep"] = imputer2.fit_transform(train_df[["CryoSleep"]])
```

```python
# train_df = train_df.drop(['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'], axis = 1)
```

```python
train_df
```

```python
X = train_df.iloc[:, :-1]
y = train_df.iloc[:, -1]
```

```python
one_hot_s = pd.get_dummies(train_df["HomePlanet"])
one_hot_d = pd.get_dummies(train_df["Destination"])

X = pd.concat([X, one_hot_s, one_hot_d], axis = 1)
```

```python
X
```

```python
X = X.drop(['Cabin', 'HomePlanet', 'Destination'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 200, criterion = 'log_loss', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report


classifier = MLPClassifier(hidden_layer_sizes=(50, 50), max_iter=500, random_state=0, 
                      solver = 'adam', activation = 'tanh',alpha = 0.01,
                     learning_rate = 'adaptive')

# Fitting the model
classifier.fit(X_train, y_train)       #0.45
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
